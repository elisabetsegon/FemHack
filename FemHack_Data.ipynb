Observation: 

There is two formats:
- 0.value
- value

We need to homogenize it. Since all "Open" have 3 relevant numbers, we can devide those bigger than 1, by 1000. So that the format is always 0.
This different in format has not corrupted the values of High and Low. So, they can be modified independently


```python
def homogenize_format(row):
    if row['Open'] > 100:
      row['Open'] = row['Open']/1000

    if row['High'] > 100:
      row['High'] = row['High']/1000

    if row['Low'] > 100:
      row['Low'] = row['Low']/1000

    if row['Close'] > 100:
      row['Close'] = row['Close']/1000
       
    return row
```

```python
def check(row):

  if row['Open'] < 0 or row['Open'] > 1:
    print("Open has not been properly adjusted", row['Open'])
    
  if row['High'] < 0 or row['High'] > 1:
    print("High has not been properly adjusted", row['High'])

  if row['Low'] < 0 or row['Low'] > 1:
    print("Low has not been properly adjusted", row['Low'])

  if row['Close'] < 0 or row['Close'] > 1:
    print("Close has not been properly adjusted", row['Close'])
```

```python
data_homogenious_format = data.apply(homogenize_format, axis=1)
```

##RSI

"The Relative Volatility Index (RVI) is similar to the Relative Strength Index (RSI) index. Both measure the direction of volatility, but RVI uses the standard deviation of price changes in its calculations, while RSI uses the absolute price changes."

"An asset is usually considered overbought when the RSI is above 70% and oversold when it is below 30%."

```python
rsi = go.Figure(data=[go.Scatter(x=data_without_outliers["Date"], y=data_without_outliers[c], name=c) for c in ['rsi']])

rsi.update_xaxes(
    rangeslider_visible=True,
    rangeselector=dict(
        buttons=list([
            dict(count=1, label="1m", step="month", stepmode="backward"),
            dict(count=6, label="6m", step="month", stepmode="backward"),
            dict(count=1, label="YTD", step="year", stepmode="todate"),
            dict(count=1, label="1y", step="year", stepmode="backward"),
            dict(step="all")
        ])
    )
)

#Adding horizontal lines at 70% and 30%
#OVERBOUGH: above 70%
x=data_without_outliers["Date"]

rsi.add_shape(type='line',
              x0=min(x),
              y0=70,
              x1=max(x),
              y1=70,
              line=dict(color='Green'))
 
#OVERSOLD: bellow 30%
rsi.add_shape(type='line',
              x0=min(x),
              y0=30,
              x1=max(x),
              y1=30,
              line=dict(color='Red')
              )

#Style
rsi.update_layout(
    title="Relative Strength Index"
)


rsi.show()
```
